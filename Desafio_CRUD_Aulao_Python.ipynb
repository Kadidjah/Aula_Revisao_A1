### Legenda das Formas (Padrão Flowgorithm)

Para interpretar o desenho acima, utilize este guia:

1.  **Elipse (Rosa):** Representa o **Início** e o **Fim** do algoritmo.
2.  **Paralelogramo (Azul):** Representa **Entrada** (Ler do teclado) e **Saída** (Escrever na tela).
3.  **Losango (Laranja):** Representa uma **Decisão** (`IF/ELSE`) ou a condição de um laço de repetição (`WHILE`). Observe que o fluxo se divide aqui.
4.  **Retângulo com borda grossa (Branco):** Representa uma **Chamada de Função**. Como seu código é modular (tem funções como `adicionar_aluno`), usamos este símbolo para manter o fluxo principal limpo.

### Detalhe da Lógica de Estatísticas

Se você der um "zoom" (duplo clique) no bloco **Chamar Estatísticas** dentro do Flowgorithm, a lógica interna seria esta:

1.  **Verificar:** A lista está vazia?
      * *Sim:* Exibir "Sem dados" $\rightarrow$ Fim da Função.
      * *Não:* Continuar.
2.  **Inicializar:** `Soma = 0`, `Maior = 0`, `Contador_Aprovados = 0`.
3.  **Loop (Para cada Aluno):**
      * `Soma = Soma + Nota`
      * `Se Nota > Maior` $\rightarrow$ `Maior = Nota`
      * `Se Nota >= 7.0` $\rightarrow$ `Contador_Aprovados = Contador_Aprovados + 1`
4.  **Cálculo Final:** `Media = Soma / Total_Alunos`.
5.  **Saída:** Exibir Média, Maior e Aprovados.



````
'''
3. Linguagem Algorítmica (Python)
Esta implementação atende aos critérios de **Lista de Dicionários**, **Modularização** e **CRUD Completo**.
'''

import os

# Variável Global para armazenar os dados (Lista de Dicionários)
# Estrutura: [{'nome': 'Ana', 'nota': 8.5}, {'nome': 'João', 'nota': 6.0}]
turma = []

def limpar_tela():
    """Limpa o console para melhorar a apresentação."""
    os.system('cls' if os.name == 'nt' else 'clear')

# --- CRUD: CREATE ---
def adicionar_aluno():
    print("--- Adicionar Aluno ---")
    nome = input("Nome do aluno: ")
    try:
        nota = float(input("Nota do aluno (0-10): "))
        if 0 <= nota <= 10:
            # Cria o dicionário e adiciona à lista
            aluno = {"nome": nome, "nota": nota}
            turma.append(aluno)
            print(f"Sucesso: {nome} adicionado.")
        else:
            print("Erro: A nota deve ser entre 0 e 10.")
    except ValueError:
        print("Erro: Digite um valor numérico para a nota.")
    input("Pressione Enter para continuar...")

# --- CRUD: READ ---
def listar_alunos():
    print("--- Lista de Alunos ---")
    if not turma:
        print("A turma está vazia.")
    else:
        print(f"{'NOME':<20} | {'NOTA':<5} | {'STATUS'}")
        print("-" * 40)
        for aluno in turma:
            status = "Aprovado" if aluno['nota'] >= 7.0 else "Reprovado"
            print(f"{aluno['nome']:<20} | {aluno['nota']:<5.1f} | {status}")
    input("Pressione Enter para continuar...")

# --- CRUD: UPDATE ---
def atualizar_nota():
    print("--- Atualizar Nota ---")
    nome_busca = input("Nome do aluno para atualizar: ")
    encontrado = False

    for aluno in turma:
        if aluno['nome'].lower() == nome_busca.lower():
            try:
                nova_nota = float(input(f"Nova nota para {aluno['nome']}: "))
                aluno['nota'] = nova_nota
                print("Nota atualizada com sucesso.")
                encontrado = True
                break
            except ValueError:
                print("Erro: Valor inválido.")

    if not encontrado:
        print("Aluno não encontrado.")
    input("Pressione Enter para continuar...")

# --- CRUD: DELETE ---
def excluir_aluno():
    print("--- Excluir Aluno ---")
    nome_busca = input("Nome do aluno para excluir: ")

    # Uso de List Comprehension para filtrar (criar nova lista sem o aluno)
    # Poderíamos usar 'remove' ou 'pop', mas filtrar é seguro contra erros de índice
    global turma
    tamanho_inicial = len(turma)
    turma = [aluno for aluno in turma if aluno['nome'].lower() != nome_busca.lower()]

    if len(turma) < tamanho_inicial:
        print("Aluno removido.")
    else:
        print("Aluno não encontrado.")
    input("Pressione Enter para continuar...")

# --- LÓGICA E ESTATÍSTICAS (O Desafio Original) ---
def calcular_estatisticas():
    print("--- Estatísticas da Turma ---")
    if not turma:
        print("Sem dados para calcular.")
    else:
        # Extrai apenas as notas para uma lista auxiliar
        notas = [aluno['nota'] for aluno in turma]

        # Cálculos
        media = sum(notas) / len(notas)
        maior_nota = max(notas)
        # Conta quantos aprovaram (nota >= 7.0)
        aprovados = sum(1 for n in notas if n >= 7.0)

        print(f"Total de Alunos: {len(turma)}")
        print(f"Média da Turma:  {media:.2f}")
        print(f"Maior Nota:      {maior_nota:.1f}")
        print(f"Aprovados:       {aprovados} ({(aprovados/len(turma))*100:.1f}%)")

    input("Pressione Enter para continuar...")

def menu():
    while True:
        limpar_tela()
        print("=== SISTEMA DE GESTÃO DE NOTAS ===")
        print("1. Adicionar Aluno")
        print("2. Listar Alunos")
        print("3. Atualizar Nota")
        print("4. Excluir Aluno")
        print("5. Relatório Estatístico (Desafio)")
        print("6. Sair")

        opcao = input("Escolha uma opção: ")

        if opcao == '1': adicionar_aluno()
        elif opcao == '2': listar_alunos()
        elif opcao == '3': atualizar_nota()
        elif opcao == '4': excluir_aluno()
        elif opcao == '5': calcular_estatisticas()
        elif opcao == '6': break
        else:
            print("Opção inválida!")
            input()

# Execução
if __name__ == "__main__":
    menu()
````

### Análise da Solução (Baseada na Ficha de Avaliação)

1.  **Lógica e Estrutura:** O código é totalmente modular. O "programa principal" é apenas um laço `while` que chama funções específicas.
2.  **Manipulação de Listas e Dicionários:** A variável `turma` é uma lista. Dentro dela, cada item é um dicionário (`{'nome': 'X', 'nota': Y}`).
3.  **Eficiência:** O uso de *List Comprehension* na função `excluir_aluno` e nas estatísticas (`sum(1 for...)`) demonstra avanço (requer domínio) da linguagem Python (Pythonic code).
4.  **Apresentação:** O uso da função `limpar_tela()` e menus formatados com `f-strings` (ex: `{media:.2f}`) garante uma experiência de usuário organizada.


````
import os  # Importa a biblioteca do sistema operacional para comandos como limpar a tela

# ---------------------------------------------------------
# VARIÁVEIS GLOBAIS
# ---------------------------------------------------------
# Lista que servirá como nosso "Banco de Dados" em memória.
# Cada item desta lista será um dicionário representando um aluno.
# Exemplo de estrutura: [{'nome': 'Ana', 'nota': 9.0}, {'nome': 'Beto', 'nota': 5.5}]
turma = []

# ---------------------------------------------------------
# FUNÇÕES AUXILIARES
# ---------------------------------------------------------
def limpar_tela():
    """
    Função responsável por limpar o console.
    Verifica o sistema operacional (Windows ou Unix/Linux/Mac)
    para executar o comando correto.
    """
    # 'nt' é o nome interno do Windows. Se for ele, usa 'cls', senão 'clear'
    os.system('cls' if os.name == 'nt' else 'clear')

def pausar():
    """Pausa a execução para que o usuário possa ler a mensagem na tela."""
    input("\nPressione [Enter] para continuar...")

# ---------------------------------------------------------
# MÓDULOS DO SISTEMA (CRUD + LÓGICA)
# ---------------------------------------------------------

def adicionar_aluno():
    """
    (CREATE) Responsável por criar um novo registro na lista.
    """
    limpar_tela()
    print("--- 1. ADICIONAR ALUNO ---")

    # Etapa 1: Coleta de dados
    nome = input("Digite o nome do aluno: ")

    # Etapa 2: Tratamento de erro (Try/Except)
    # Tentamos converter a entrada para float. Se falhar, capturamos o erro.
    try:
        nota = float(input("Digite a nota do aluno (0 a 10): "))

        # Etapa 3: Validação de regra de negócio
        if 0 <= nota <= 10:
            # Cria um dicionário com os dados validados
            novo_aluno = {
                "nome": nome,
                "nota": nota
            }

            # Adiciona o dicionário à lista principal (append)
            turma.append(novo_aluno)
            print(f"\nSucesso! Aluno '{nome}' cadastrado com nota {nota}.")
        else:
            print("\nErro: A nota deve ser um valor entre 0 e 10.")

    except ValueError:
        # Executado se o usuário digitar letras ao invés de números na nota
        print("\nErro: Por favor, digite apenas números válidos para a nota (ex: 8.5).")

    pausar()

def listar_alunos():
    """
    (READ) Responsável por ler e exibir os dados armazenados.
    """
    limpar_tela()
    print("--- 2. LISTA DE ALUNOS ---")

    # Verifica se a lista está vazia antes de tentar processar
    if not turma:
        print("A turma está vazia. Nenhum aluno cadastrado.")
    else:
        # Cabeçalho da tabela com formatação de espaçamento
        print(f"{'NOME':<20} | {'NOTA':<6} | {'SITUAÇÃO'}")
        print("-" * 45) # Linha separadora visual

        # Etapa de Iteração: Percorre cada dicionário na lista
        for aluno in turma:
            nome_atual = aluno['nome']
            nota_atual = aluno['nota']

            # Lógica condicional em linha (Operador Ternário) para definir status
            situacao = "APROVADO" if nota_atual >= 7.0 else "REPROVADO"

            # Exibe os dados formatados
            # :<20 alinha à esquerda com 20 espaços
            # :.1f formata o número float com 1 casa decimal
            print(f"{nome_atual:<20} | {nota_atual:<6.1f} | {situacao}")

    pausar()

def atualizar_nota():
    """
    (UPDATE) Responsável por alterar um dado existente.
    """
    limpar_tela()
    print("--- 3. ATUALIZAR NOTA ---")

    # Coleta o nome de busca
    nome_busca = input("Qual aluno deseja alterar? ")
    aluno_encontrado = False # Variável de controle (flag)

    # Percorre a lista procurando pelo nome
    for aluno in turma:
        # Compara os nomes em minúsculo (lower) para evitar erros de digitação (Ana vs ana)
        if aluno['nome'].lower() == nome_busca.lower():
            print(f"Aluno encontrado: {aluno['nome']} (Nota atual: {aluno['nota']})")

            try:
                nova_nota = float(input("Digite a NOVA nota: "))
                if 0 <= nova_nota <= 10:
                    # Atualiza o valor da chave 'nota' no dicionário
                    aluno['nota'] = nova_nota
                    print("\nNota atualizada com sucesso!")
                    aluno_encontrado = True
                    break # Para o laço pois já encontramos o aluno
                else:
                    print("Erro: Nota fora do intervalo 0-10.")
            except ValueError:
                print("Erro: Valor não numérico.")

            # Se houve erro de validação dentro do loop, marcamos que encontramos o aluno
            # mas a atualização falhou. O break acima só ocorre no sucesso.
            aluno_encontrado = True
            break

    if not aluno_encontrado:
        print("\nAluno não encontrado na base de dados.")

    pausar()

def excluir_aluno():
    """
    (DELETE) Responsável por remover um registro da lista.
    """
    limpar_tela()
    print("--- 4. EXCLUIR ALUNO ---")

    nome_busca = input("Nome do aluno a ser excluído: ")

    # Precisamos referenciar a variável global pois vamos reescrevê-la
    global turma

    # Guarda o tamanho atual para verificar depois se algo foi removido
    tamanho_antes = len(turma)

    # --- TÉCNICA DE FILTRAGEM (List Comprehension) ---
    # Recria a lista 'turma' mantendo apenas quem tem o nome DIFERENTE do buscado.
    # Isso efetivamente "apaga" quem tiver o nome igual.
    turma = [
        aluno for aluno in turma
        if aluno['nome'].lower() != nome_busca.lower()
    ]

    tamanho_depois = len(turma)

    # Verifica se o tamanho diminuiu (significa que alguém foi removido)
    if tamanho_depois < tamanho_antes:
        print(f"\nAluno '{nome_busca}' removido com sucesso.")
    else:
        print("\nAluno não encontrado. Ninguém foi excluído.")

    pausar()

def estatisticas_turma():
    """
    (LÓGICA EXTRA) Realiza cálculos matemáticos sobre os dados.
    """
    limpar_tela()
    print("--- 5. ESTATÍSTICAS DA TURMA ---")

    if not turma:
        print("Não há dados suficientes para gerar estatísticas.")
    else:
        # Cria uma lista simples contendo apenas as notas (ex: [8.0, 5.5, 9.0])
        lista_notas = [aluno['nota'] for aluno in turma]

        # 1. Cálculo da Média (Soma das notas / Quantidade de notas)
        media = sum(lista_notas) / len(lista_notas)

        # 2. Maior Nota (Função max)
        maior_nota = max(lista_notas)

        # 3. Contagem de Aprovados
        # Gera 1 para cada nota >= 7 e soma tudo
        qtd_aprovados = sum(1 for nota in lista_notas if nota >= 7.0)

        # 4. Cálculo da Porcentagem
        porcentagem_aprov = (qtd_aprovados / len(turma)) * 100

        print(f"Total de Alunos:  {len(turma)}")
        print(f"Média da Turma:   {media:.2f}")
        print(f"Maior Nota:       {maior_nota:.1f}")
        print(f"Total Aprovados:  {qtd_aprovados} alunos ({porcentagem_aprov:.1f}%)")

    pausar()

# ---------------------------------------------------------
# PROGRAMA PRINCIPAL (MENU)
# ---------------------------------------------------------
def menu_principal():
    """
    Função controladora que exibe o menu e gerencia o fluxo.
    """
    while True: # Loop Infinito (roda até encontrar um 'break')
        limpar_tela()
        print("="*30)
        print("   SISTEMA DE GESTÃO ESCOLAR")
        print("="*30)
        print("1. Adicionar Aluno")
        print("2. Listar Alunos")
        print("3. Atualizar Nota")
        print("4. Excluir Aluno")
        print("5. Ver Estatísticas")
        print("6. Sair do Sistema")
        print("-" * 30)

        opcao = input("Escolha uma opção: ")

        # Estrutura de decisão para chamar a função correta
        if opcao == '1':
            adicionar_aluno()
        elif opcao == '2':
            listar_alunos()
        elif opcao == '3':
            atualizar_nota()
        elif opcao == '4':
            excluir_aluno()
        elif opcao == '5':
            estatisticas_turma()
        elif opcao == '6':
            print("Encerrando o sistema...")
            break # Quebra o While e encerra o programa
        else:
            print("Opção inválida!")
            pausar()

# Verifica se o script está sendo executado diretamente
if __name__ == "__main__":
    menu_principal()
````
